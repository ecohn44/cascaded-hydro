```r
## DATA LOAD

library(readxl)
library(ggplot2)
```

```r
# Lower Monumental Data Processing

LMN_A <- read_excel("streamflow-data/LMN6A_daily.xlsx", sheet = 1)  
```

```r
colnames(LMN_A)[colnames(LMN_A) == "A (unit:cfs)"] <- "inflow"
```

```r
summary(LMN_A)
```

```r
ggplot(LMN_A, aes(x = date, y = inflow, group = 1)) + geom_line()+labs(title = "Lower Monumental Inflow 1928-2018", x = "Date", y = "Inflow (cfs)") 
```

```r
# Lower Granite Data Processing

LWG_A <- read_excel("streamflow-data/LWG6A_daily.xlsx", sheet = 1)  
colnames(LWG_A)[colnames(LWG_A) == "A (unit:cfs)"] <- "inflow"
```

```r
# Little Goose Data Processing
LGS_H <- read_excel("streamflow-data/LGS6H_daily.xlsx", sheet = 1)  
colnames(LGS_H)[colnames(LGS_H) == "H (unit:cfs)"] <- "outflow"
```

```r
LGS_S <- read_excel("streamflow-data/LGS6S_daily.xlsx", sheet = 1)  
colnames(LGS_S)[colnames(LGS_S) == "S (unit:cfs)"] <- "storage"
```

```r
# Merge by date
LGS_A <- merge(LGS_H, LGS_S, by = "date", suffixes = c("_H", "_S"))

LGS_A$inflow <- LGS_A$outflow + LGS_A$storage
```

```r
ggplot(LGS_A, aes(x = date, y = inflow, group = 1)) + geom_line()+labs(title = "Little Goose Inflow 1928-2018", x = "Date", y = "Inflow (cfs)") 
```
