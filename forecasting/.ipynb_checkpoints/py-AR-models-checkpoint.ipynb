## Data Import & Visualization

```python
## Lower Monumental

LMN = pd.read_excel("streamflow-data/LMN6A_daily.xlsx")
LMN = LMN.rename(columns={'A (unit:cfs)': 'inflow'})
LMN['log_inflow'] = np.log(LMN['inflow'])
```

```python
## Lower Granite 

LWG = pd.read_excel("streamflow-data/LWG6A_daily.xlsx")
LWG = LWG.rename(columns={'A (unit:cfs)': 'inflow'})
LWG['log_inflow'] = np.log(LWG['inflow'])
```

```python
## Little Goose 

LGS_H = pd.read_excel("streamflow-data/LGS6H_daily.xlsx")  
LGS_H = LGS_H.rename(columns={'H (unit:cfs)': 'outflow'})

LGS_S = pd.read_excel("streamflow-data/LGS6S_daily.xlsx")  
LGS_S = LGS_S.rename(columns={'S (unit:cfs)': 'storage'})

# Merge by date
LGS = pd.merge(LGS_H, LGS_S,  how='inner', left_on=['date'], right_on = ['date'])

# Calculate inflow
LGS['inflow'] = LGS['outflow'] + LGS['storage']
LGS['log_inflow'] = np.log(LGS['inflow'])
```

```python
## Ice Harbor

IHR_H = pd.read_excel("streamflow-data/IHR6H_daily.xlsx")  
IHR_H = IHR_H.rename(columns={'H (unit:cfs)': 'outflow'})

IHR_S = pd.read_excel("streamflow-data/IHR6S_daily.xlsx")  
IHR_S = IHR_S.rename(columns={'S (unit:cfs)': 'storage'})

# Merge by date
IHR = pd.merge(IHR_H, IHR_S,  how='inner', left_on=['date'], right_on = ['date'])

# Calculate inflow
IHR['inflow'] = IHR['outflow'] + IHR['storage']
IHR['log_inflow'] = np.log(IHR['inflow'])
```

### Dataset Creation: 365 x 4 Time Series over 100 Years

```python
def pivot_data(df):
    # Ensure the DateTime index
    df['date'] = pd.to_datetime(df['date'])
    df.set_index('date', inplace=True)
    
    # Extract year and day-of-year
    df['Year'] = df.index.year
    df['DayOfYear'] = df.index.dayofyear
    
    # Remove leap year extra days (day 366)
    df = df[df['DayOfYear'] <= 365]

    # Filter out rows before 1929 and after 2017
    df = df.loc['1929-01-01':'2017-12-31']

    # Pivot the data: rows = years, columns = days of the year
    df_pivot = df.pivot(index='Year', columns='DayOfYear', values='log_inflow')

    return df_pivot
```

```python
LMN_pivot = pivot_data(LMN)
LWG_pivot = pivot_data(LWG)
LGS_pivot = pivot_data(LGS)
IHR_pivot = pivot_data(IHR)
```

```python
from datetime import datetime, timedelta

def dt_convert(day_of_year, year=2022):  # Not leap year
    date = datetime(year, 1, 1) + timedelta(days=day_of_year - 1)
    return date.strftime("%B %d")  # Returns month name and day
```

## AR(p) Process

```python
## Lets try to build an AR model for one day of the year...

# Split Train / Test Data
p = 10
n = len(LMN_pivot) # number of years available 
#test_n = 10 #[1, 5, 10]
#train_n = n - test_n

# Day 1 
d = 1
data = LMN_pivot[d].to_frame()
```

```python
# Adding lag features to the DataFrame
for i in range(1, p):  # Creating lag features up to p days
    data[f'Lag_{i}'] = data[d].shift(i)

# Drop rows with NaN values resulting from creating lag features
data.dropna(inplace=True)
data = data.rename(columns={1: '1'})
d = str(d)

# Split the data into training and testing sets
train_size = int(0.8 * len(data))
train_data = data[:train_size]
test_data = data[train_size:]

# Define the input features (lag features) and target variable
y_train = train_data[d].reset_index(drop=True)
y_train.name = None
y_test = test_data[d].reset_index(drop=True)
y_test.name = None
```

```python
from statsmodels.graphics.tsaplots import plot_acf

series = data[d]
plot_acf(series)
plt.show()
```

```python
# Define the number of future time steps you want to predict (next 7 steps)
forecast_steps = 7

# Extend the predictions into the future for one year
future_indices = range(len(test_data), len(test_data) + forecast_steps)
future_predictions = ar_results.predict(start=len(train_data), end=len(train_data) + len(test_data) + forecast_steps - 1, dynamic=False)

# Create date indices for the future predictions
future_dates = [test_data.index[-1] + i for i in range(1, forecast_steps+1)]
```
